## Handling the outliers using the quantile method

```python
# 10th quantile and 90th quantile identification for outlier treatment
print(df['Wind Speed (km/h)'].quantile(0.10))
print(df['Wind Speed (km/h)'].quantile(0.90))
```

```python
# OUTLIER TREATMENT!!!
df["Wind Speed (km/h)"] = np.where(df["Wind Speed (km/h)"] <3.2039, 3.2039,df['Wind Speed (km/h)'])
df["Wind Speed (km/h)"] = np.where(df["Wind Speed (km/h)"] >20.4148, 20.4148,df['Wind Speed (km/h)'])
print(df['Wind Speed (km/h)'].skew())
```

```python
# 10th quantile and 90th quantile identification for outlier treatment
print(df['Humidity'].quantile(0.10))
print(df['Humidity'].quantile(0.90))
```

```python
# OUTLIER TREATMENT!!!
df["Humidity"] = np.where(df["Humidity"] <0.43, 0.43,df['Humidity'])
df["Humidity"] = np.where(df["Humidity"] >0.95, 0.95,df['Humidity'])
print(df['Humidity'].skew())
```

```python
# 10th quantile and 90th quantile identification for outlier treatment
print(df['Pressure (millibars)'].quantile(0.10))
print(df['Pressure (millibars)'].quantile(0.90))
```

```python
# OUTLIER TREATMENT!!!
df["Pressure (millibars)"] = np.where(df["Pressure (millibars)"] <1007.07, 1007.07,df['Pressure (millibars)'])
df["Pressure (millibars)"] = np.where(df["Pressure (millibars)"] >1026.61, 1026.61,df['Pressure (millibars)'])
print(df['Pressure (millibars)'].skew())
```

```python
# 10th quantile and 90th quantile identification for outlier treatment
print(df['Apparent Temperature (C)'].quantile(0.10))
print(df['Apparent Temperature (C)'].quantile(0.90))
```

```python
# OUTLIER TREATMENT!!!
df["Apparent Temperature (C)"] = np.where(df["Apparent Temperature (C)"] <-3.311, -3.311,df['Apparent Temperature (C)'])
df["Apparent Temperature (C)"] = np.where(df["Apparent Temperature (C)"] >24.9055, 24.9055,df['Apparent Temperature (C)'])
print(df['Apparent Temperature (C)'].skew())
```
